### Quest√µes 3 e 4

```python
def young(a):
    return m.exp(-a**2/800)

def old(a):
    return m.exp(-(a-100)**2/1800)
```

```python
def very_young(a):
    return m.exp(-a**2/400)

def very_old(a):
    return m.exp(-(a-100)**2/900)
```

```python
def not_very_young(a):
    return 1 - m.exp(-a**2/400)

def not_very_old(a):
    return 1 - m.exp(-(a-100)**2/900)
```

```python
def new_very_young(a):
    return m.exp(-a**2/800)**2

def new_very_old(a):
    return m.exp(-(a-100)**2/1800)**2
```

```python
x = np.linspace(0, 100, num=500)
y1 = np.vectorize(very_young)(x)
y2 = np.vectorize(very_old)(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
x = np.linspace(0, 100, num=500)
y1 = np.vectorize(not_very_young)(x)
y2 = np.vectorize(not_very_old)(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

## Quest√£o 6

```python
def trimf(x, a, b, c):
    y = np.empty(x.size)
    for i in range(x.size):
        y[i] = (max(min( 1.0*(x[i]-a)/(b-a), 1.0*(c-x[i])/(c-b)),0))
    
    return y

def trapmf(x, a, b, c, d):
    y = np.empty(x.size)
    for i in range(x.size):
        y[i] = (max(min( 1.0*(x[i]-a)/(b-a), 1, 1.0*(d-x[i])/(d-c) ), 0))
        
    return y
```

```python
x = np.linspace(0, 10, num=500)

mu_a1 = trapmf(x, 3, 4, 5, 6)
mu_a2 = trapmf(x, 6, 6.5, 7, 7.5)
mu_aline = trimf(x, 5, 6, 7)
max_a1line = max(np.minimum(mu_a1, mu_aline))
max_a2line = max(np.minimum(mu_a2, mu_aline))
print(f"M√°ximo entre A1 e A': {max_a1line}")
print(f"M√°ximo entre A2 e A': {max_a2line}")
plt.plot(x, mu_a1, color='blue')
plt.plot(x, mu_a2, color='orange')
plt.plot(x, mu_aline, color='green')
plt.plot(x, np.ones(x.size)*max_a1line, ls='--')
plt.plot(x, np.ones(x.size)*max_a2line, ls='--')
plt.show()

mu_c1 = trimf(x, 3, 4, 5)
mu_c2 = trimf(x, 4, 5, 6)
plt.plot(x, mu_c1)
plt.plot(x, mu_c2)
plt.show()
```

```python
min_comp = min(max_a1line, max_a2line)
min_comp_array = np.ones(x.size)*min_comp

mu_c1 = trimf(x, 3, 4, 5)
mu_c2 = trimf(x, 4, 5, 6)
plt.plot(x, np.minimum(mu_c1, min_comp_array))
plt.plot(x, np.minimum(mu_c2, min_comp_array))
plt.ylim(0, 1)
plt.show()
```

```python
mu_cline = np.maximum(np.minimum(mu_c1, min_comp_array), np.minimum(mu_c2, min_comp_array))
plt.plot(x, mu_cline)
plt.ylim(0, 1)
plt.show()
```

## Aproxima√ß√£o da fun√ß√£o cosseno

Considerando as seguintes regras dentro do intervalo $\left[0, 2\pi\right]$: 

* Se x √© pequeno, ent√£o $y = \cfrac{2x}{\pi} + 1$
* Se x √© m√©dio, ent√£o $y = \cfrac{-2x}{\pi} + 1$
* Se x √© grande, ent√£o $y = \cfrac{2x}{\pi} - 3$

```python
def y_small(x):
    return x*2/m.pi+1

def y_medium(x):
    return -x*2/m.pi+1

def y_big(x):
    return x*2/m.pi-3

# python >= 3.10
'''
def f(x, member):
    match member:
        case 'small':
            return x*2/m.pi+1
        case 'big small':
            return 1
        case 'medium':
            return -x*2/m.pi+1
        case 'big medium':
            return -1
        case 'big':
            return x*2/m.pi-3
        case _:
            return 0
'''

# python < 3.10     
# tem que criar as fun√ß√µes para cada classe, que agora s√£o 7
def f(x, member):
    if (member == 'small'):
        return x*2/m.pi+1
    elif (member == 'big small'):
        return 1
    elif (member == 'medium'):
        return -x*2/m.pi+1
    elif (member == 'big medium'):
        return -1
    elif (member == 'big'):
        return x*2/m.pi-3
    else:
        return 0
    
'''
def y_takagi_sugeno(x_small, x_medium, x_big):
    if (x_small + x_medium + x_big == 0):
        return 0
    return (x_small*y_small(x_small) + x_medium*y_medium(x_medium) + x_big*y_big(x_big))#/(x_small+x_medium+x_big)
'''

# Esse x tem members.size elementos e cont√©m o pertencimento de um dado x a cada uma das classes
def y_takagi_sugeno(x, members):
    
    if (x.sum() == 0):
        return 0
    
    total = 0
    for i in range(x.size):
        total = total + x[i]*f(i, members[i])
        
    return total
```

```python
x = np.linspace(-m.pi/2, 3*m.pi/2, num=500)
members = ['small', 'big small', 'medium', 'big medium', 'big']

mu = np.empty(members.size)

start = -m.pi/2
step = m.pi/2
for i in range(members.size):    
    mu[i] = trimf(x, start, start+step, start+2*step)
    start = start + m.pi/4

plt.plot(x, np.vectorize(m.cos)(x), color='black')
plt.plot(x, np.vectorize(y_small)(x), color='blue')
plt.plot(x, np.vectorize(y_medium)(x), color='orange')
plt.plot(x, np.vectorize(y_big)(x), color='red')
plt.plot(x, mu_small-2, color='blue')
plt.plot(x, mu_medium-2, color='orange')
plt.plot(x, mu_big-2, color='red')
plt.ylim(-2, 1.25)
plt.show()
```

```python
x = np.linspace(-m.pi/2, 3*m.pi/2, num=500)
members = np.array(['very small',
                    'small',
                    'neither small or medium',
                    'medium',
                    'neither medium or big',
                    'big',
                    'very big'])

mu = np.empty((members.size, x.size))

start = -m.pi/2
step = m.pi/2
for i in range(members.size):    
    mu[i] = trimf(x, start, start+step, start+2*step)
    start = start + m.pi/8
    plt.plot(x, mu[i])

plt.show()
```

```python
y = np.empty(x.size)

for i in range(x.size):
    membership = np.empty(members.size)
    for j in range(members.size):
        membership[j] = mu[j][i]
    y[i] = y_takagi_sugeno(membership, members)

plt.plot(x, y)
plt.show()
```
