```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

start = 0
end = 10
x = np.linspace(start, end, 100)
print(x)
```

## Função triangular

```python
def trimf(x, a, b, c):
    y = []
    for num in x:
        y.append(max(min( 1.0*(num-a)/(b-a), 1.0*(c-num)/(c-b)),0))
    
    return y
```

```python
y = trimf(x, 3, 6, 8)

plt.plot(x, y, linewidth=2.0)
x_range = np.arange(start, end + 0.1, 1)
y_range = np.arange(min(min(y),0), max(y) + 0.1, 0.1)
plt.xticks(x_range)
plt.yticks(y_range)

plt.show()
```

## Função trapézio

```python
def trapmf(x, a, b, c, d):
    y = []
    for num in x:
        y.append(max(min( 1.0*(num-a)/(b-a), 1, 1.0*(d-num)/(d-c) ), 0))
        
    return y
```

```python
y = trapmf(x, 1, 5, 7, 8)

plt.plot(x, y, linewidth=2.0)
x_range = np.arange(start, end + 0.1, 1)
y_range = np.arange(min(min(y),0), max(y) + 0.1, 0.1)
plt.xticks(x_range)
plt.yticks(y_range)

plt.show()
```

## Função gaussiana

```python
def gaussmf(x, c, sigma):
    y = []
    for num in x:
        y.append( exp( -1.0*(num-c)**2/(2*sigma**2) ) )
        
    return y
```

```python
y = gaussmf(x, 5, 2)

plt.plot(x, y, linewidth=2.0)
x_range = np.arange(start, end + 0.1, 1)
y_range = np.arange(min(min(y),0), max(y) + 0.1, 0.1)
plt.xticks(x_range)
plt.yticks(y_range)

plt.show()
```

## Função sino generalizada

```python
def gbellmf(x, a, b, c):
    y = []
    for num in x:
        y.append( 1.0/(1 + abs((num - c)/a)**(2*b) ) )
        
    return y
```

```python
y = gbellmf(x, 2, 4, 6)

plt.plot(x, y, linewidth=2.0)
x_range = np.arange(start, end + 0.1, 1)
y_range = np.arange(min(min(y),0), max(y) + 0.1, 0.1)
plt.xticks(x_range)
plt.yticks(y_range)

plt.show()
```

## Função sigmoidal

```python
def sigmf(x, a, b):
    y = []
    for num in x:
        y.append( 1.0/(1+exp(-a*(num-b))) )
    
    return y
```

```python
y = sigmf(x, 2, 4)

plt.plot(x, y, linewidth=2.0)
x_range = np.arange(start, end + 0.1, 1)
y_range = np.arange(min(min(y),0), max(y) + 0.1, 0.1)
plt.xticks(x_range)
plt.yticks(y_range)

plt.show()
```
