## Funções

### Cria os centros

Essa função cria o vetor de centroides do K-means com valores aleatórios com base no número inicial de centroides e no número de canais. Aqui a função `default_rng(seed).random((a,b))` produz uma matriz de *a* linhas por *b* colunas onde cada elemento está no intervalo $[0,1[$. O número de colunas é definido como 3, pois só há três canais de cor: vermelhor, verde e azul (RGB). A função `vectorize` recebe uma função como argumento e retorna uma outra função que aplica a função argumento no array que lhe é passado. Ou seja, aqui, a parte `vectorize(floor)` cria uma função que aplica a função piso em um array. O que se segue é uma passagem de argumento, no caso, uma matriz de números decimais que precisam ser arredondados.

```python
def make_centers_array(n, channels=3, seed=0):
    return vectorize(floor)(default_rng(seed).random((n, channels))*256)
```

### Importação e vetorização da imagem

Essa função importa a imagem que inicialmente é um tensor de 3 dimensões, ou seja, uma série de matrizes empilhadas. Porém, a manipulação é mais fácil se as imagens forem tratadas como um grande vetor de pixels, onde cada pixel é representado por três valores: um para cada canal e é isso que esta função se empenha em fazer. O formato inicial da imagem é guardado para usos futuros.

```python
def make_img_array(name, display=True, channels=3):
    image = Image.open(name)
    
    if (display):
        display(image)
    
    array = asarray(image)
    shape = array.shape
    array = array.reshape(1, array.size // channels, channels)[0]
    
    return (array, shape)
```

### Distância de cada ponto ao centro

A primeira função calcula a distância entre um dado pixel e os centros definidos no `make_img_array`. A segunda, calcula o pertencimento de cada pixel à classe que cada centro representa.

```python
def distance_from_pixel_to_centers(pixel, centers):
    
    n = centers.shape[0]
    dist = empty(shape=(n))
    
    for i in range(n):
        
        dist[i] = norm(pixel - centers[i])
        
    return dist
```

```python
def start_belonging_array(img_2d_array, centers):
    
    n = centers.shape[0]
    n_pixels = img_2d_array.shape[0]
    belonging = default_rng().random((n_pixels, 1))
    
    for i in range(n-1):
        
        if (i == n - 2):
            belongings = concatenate([belonging, 1 - belongings.sum(axis=1)], axis=1)
            break
        
        next_belonging_column = default_rng().random((n_pixels, 1)) * (1 - belongings.sum(axis=1))
        belongings = concatenate([belonging, next_belonging_column], axis=1)
        
    return belonging
        
```

```python
def belonging_array(img_2d_array, centers):
    
    n = centers.shape[0]
    n_pixels = img_2d_array.shape[0]
    dist = empty(shape=(n_pixels, n))
    belonging = empty(shape=(n_pixels, n))
    
    for index in range(n_pixels):
        dist[index] = distance_from_pixel_to_centers(img_2d_array[index], centers)
        
    squared_dist = dist**2
    squared_dist_sum = squared_dist.sum(axis=1)
    #print(f'dist =\n{dist[:3]}\nsquared dist = {squared_dist[:3]}\nsquared dist sum = {squared_dist_sum[:3]}\n')
        
    for index in range(n_pixels):
        belonging[index] = 1 / ( squared_dist[index] / squared_dist_sum[index] )**2
        
    #print(f'\n\nbelonging = {belonging[:3]}\ndist[-3:] = {dist[-3:]}')
    
    return belonging
```

### Atualização dos centroides

A função abaixo, simplesmente atualiza os centros com base nos pertencimentos e pixels da imagem. Ela também devolve a soma das distâncias entre os centros novos e antigos.

```python
def center_refresh(img_2d_array, centers, belonging):
    
    n = centers.shape[0]
    n_pixels = img_2d_array.shape[0]
    new_centers = empty(shape=centers.shape)
    changing = empty(n)
    
    for i in range(n):
        
        print(f'3 pixels = {img_2d_array[:3]} \n3 belongings = {belonging[:3]}')
        sum_with_weight = ((belonging[:,i]**2).reshape([n_pixels, 1]) * img_2d_array).sum(axis=0)
        sum_no_weight = (belonging[:,i]**2).sum(axis=0)
        
        new_centers[i] = sum_with_weight / sum_no_weight
            
        changing[i] = norm(new_centers[i] - centers[i])
        print(f'sum with weight = {sum_with_weight} \nsum no weight = {sum_no_weight}\nNew center = {new_centers[i]}')
    
    return (new_centers, changing.sum())        
```

### Convergência dos centroides

```python
n_lake = 2
n_flower = 3
epsilon = 0.001
max_epochs = 50
changing_lake = epsilon + 1
epochs = 0

centers_lake = make_centers_array(n_lake, seed=5)
centers_flower = make_centers_array(n_flower, seed=1)

array_lake, shape_lake = make_img_array('photo010.jpg', display=False)
array_flower, shape_flower = make_img_array('photo003.jpg', display=False)

print(centers_lake)

while((changing_lake > epsilon) and (epochs < max_epochs)):
    belonging_lake = belonging_array(array_lake, centers_lake)
    print(belonging_lake)
    centers_lake, changing_lake = center_refresh(array_lake, centers_lake, belonging_lake)
    epochs = epochs + 1
    
    if (changing_lake <= epsilon):
        print(f'Terminou por epsilon. Épocas = {epochs}')
    if (epochs >= max_epochs):
        print(f'Terminou por épocas. Épsilon = {changing_lake}')
    
    if (epochs % 7 == 0):
        print(f'centers = {centers_lake}\nchanging = {changing_lake}')
```
